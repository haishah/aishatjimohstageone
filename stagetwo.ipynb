```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```
url = 'https://archive.ics.uci.edu/ml/machine-learning-databases/00374/energydata_complete.csv'
df = pd.read_csv(url)
df.head()
```

```
df = df.drop(['date','lights'], axis=1)
df
```

```
from sklearn.preprocessing import MinMaxScaler
scalar = MinMaxScaler()
normalised_df = pd.DataFrame(scalar.fit_transform(df), columns = df.columns)
df_normalised = normalised_df.drop(columns = ['Appliances'])
df_tv = normalised_df['Appliances'] 
```

```
#spliting the datasets
from sklearn.model_selection import train_test_split
X_train, X_val, y_train, y_val = train_test_split(df_normalised, df_tv, test_size = 0.3, random_state = 42)
#Define model
from sklearn.linear_model import LinearRegression
linear_model = LinearRegression()
#fit the model to the training dataset
linear_model.fit(X_train,y_train) 
```

```
from sklearn.metrics import mean_absolute_error
predicted_values = linear_model.predict(X_val)
mse = mean_absolute_error(y_val,predicted_values)
print(mse.round(2))
```

```
rss = np.sum(np.square(y_val - predicted_values))
round(rss,2)
```

```
from sklearn.metrics import mean_squared_error
rmse = np.sqrt(mean_squared_error(y_val, predicted_values))
print(rmse.round(3))
```

```
from sklearn.metrics import r2_score
r2_score = r2_score(y_val, predicted_values)
round(r2_score, 2)
```
